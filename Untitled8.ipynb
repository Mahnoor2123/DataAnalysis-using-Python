```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Example data
data = np.array([[2.5, 2.4],
                 [0.5, 0.7],
                 [2.2, 2.9],
                 [1.9, 2.2],
                 [3.1, 3.0],
                 [2.3, 2.7],
                 [2, 1.6],
                 [1, 1.1],
                 [1.5, 1.6],
                 [1.1, 0.9]])

# Standardize the data
scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)

# Perform PCA
pca = PCA(n_components=2)  # Number of components
principal_components = pca.fit_transform(data_standardized)

# Extract PC1 and PC2
pc1 = principal_components[:, 0]  # First principal component
pc2 = principal_components[:, 1]  # Second principal component

print("Principal Component 1:\n", pc1)
print("Principal Component 2:\n", pc2)

# Variance explained by PC1 and PC2
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
data = iris.data  # Features
target = iris.target  # Target (labels)

# Create a DataFrame for better visualization
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target

# View the first few rows
print(df.head())
```

```python
# Standardize the features (ignore the target column)
scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)
```

```python
# Perform PCA and select the first 2 components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_standardized)

# Add the principal components to the DataFrame
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca['target'] = target

# View the first few rows
print(df_pca.head())
```

```python
import matplotlib.pyplot as plt

# Plot the PCA-transformed data
plt.figure(figsize=(8, 6))
for label, color in zip([0, 1, 2], ['red', 'green', 'blue']):
    plt.scatter(
        df_pca[df_pca['target'] == label]['PC1'],
        df_pca[df_pca['target'] == label]['PC2'],
        label=iris.target_names[label],
        color=color
    )
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Iris Dataset')
plt.legend()
plt.grid()
plt.show()
```

```python
# Variance explained by each principal component
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```
